```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# Parmetros de la distribucin lognormal
mu = 2  # Media en escala logartmica
sigma = 0.48  # Desviacin estndar en escala logartmica

# Crear un rango de valores de corriente (kA)
x = np.linspace(0.1, 30, 1000)

# Calcular la PDF y CDF
pdf = lognorm.pdf(x, sigma, scale=np.exp(mu))
cdf = lognorm.cdf(x, sigma, scale=np.exp(mu))

# Crear la figura y los ejes
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

# Graficar la PDF
ax1.plot(x, pdf, 'b-', label="Probabilidad de densidad [PDF]")
ax1.set_xlabel("Corriente (kA)")
ax1.set_ylabel("Densidad de Probabilidad", color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Graficar la CDF
ax2.plot(x, cdf, 'orange', label="Probabilidad Acumulada [CDF]")
ax2.set_ylabel("Probabilidad Acumulada", color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Agregar ttulo y leyenda
title = "Lognormal de la Corriente de Rayos"
plt.title(title)
ax1.legend(loc="lower right")
ax2.legend(loc="upper left")

# Mostrar la cuadrcula
ax1.grid(True)

# Mostrar la grfica
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# Parmetros de la distribucin lognormal
mu = 2  # Media en escala logartmica
sigma = 0.48  # Desviacin estndar en escala logartmica

# Crear un rango de valores de corriente (kA)
x = np.linspace(0.1, 30, 1000)

# Calcular la PDF y CDF
pdf = lognorm.pdf(x, sigma, scale=np.exp(mu))
cdf = lognorm.cdf(x, sigma, scale=np.exp(mu))

# Crear la figura y los ejes
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

# Graficar la PDF con nuevo color
ax1.plot(x, pdf, 'green', label="Probabilidad de densidad [PDF]")
ax1.set_xlabel("Corriente (kA)")
ax1.set_ylabel("Densidad de Probabilidad", color='green')
ax1.tick_params(axis='y', labelcolor='green')

# Graficar la CDF con nuevo color
ax2.plot(x, cdf, 'purple', label="Probabilidad Acumulada [CDF]")
ax2.set_ylabel("Probabilidad Acumulada", color='purple')
ax2.tick_params(axis='y', labelcolor='purple')

# Agregar ttulo y leyenda
title = "Lognormal de la Corriente de Rayos"
plt.title(title)
ax1.legend(loc="lower right")
ax2.legend(loc="upper left")

# Mostrar la cuadrcula
ax1.grid(True)

# Mostrar media y desviacin estndar en la grfica
plt.text(15, 0.1, f"$\mu$ = {mu}, $\sigma$ = {sigma}", fontsize=12, bbox=dict(facecolor='white', alpha=0.6))

# Permitir seleccionar un nivel de corriente y graficarlo
selected_current = 10  # Se puede cambiar a cualquier valor deseado
pdf_selected = lognorm.pdf(selected_current, sigma, scale=np.exp(mu))
cdf_selected = lognorm.cdf(selected_current, sigma, scale=np.exp(mu))
ax1.axvline(selected_current, color='blue', linestyle='--', label=f"Corriente = {selected_current} kA")
ax1.scatter([selected_current], [pdf_selected], color='blue', zorder=3)
ax2.scatter([selected_current], [cdf_selected], color='blue', zorder=3)
ax1.legend()
ax2.legend()

# Mostrar la grfica
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

# Parmetros de la distribucin lognormal
mu = 2  # Media en escala logartmica
sigma = 0.48  # Desviacin estndar en escala logartmica

# Crear un rango de valores de corriente (kA)
x = np.linspace(0.1, 30, 1000)

# Calcular la PDF y CDF
pdf = lognorm.pdf(x, sigma, scale=np.exp(mu))
cdf = lognorm.cdf(x, sigma, scale=np.exp(mu))

# Crear la figura y los ejes
fig, ax1 = plt.subplots(figsize=(10, 6))
ax2 = ax1.twinx()

# Graficar la PDF con nuevo color
ax1.plot(x, pdf, 'green', label="Probabilidad de densidad [PDF]")
ax1.set_xlabel("Corriente (kA)", fontsize=14)
ax1.set_ylabel("Densidad de Probabilidad", color='green', fontsize=14)
ax1.tick_params(axis='y', labelcolor='green', labelsize=12)
ax1.tick_params(axis='x', labelsize=12)

# Graficar la CDF con nuevo color
ax2.plot(x, cdf, 'purple', label="Probabilidad Acumulada [CDF]")
ax2.set_ylabel("Probabilidad Acumulada", color='purple', fontsize=14)
ax2.tick_params(axis='y', labelcolor='purple', labelsize=12)

# Agregar ttulo y leyenda
title = "Lognormal de la Corriente de Rayos"
plt.title(title, fontsize=16)
ax1.legend(loc="upper left", fontsize=12)
ax2.legend(loc="lower right", fontsize=12)

# Mostrar la cuadrcula
ax1.grid(True)

# Mostrar media y desviacin estndar en la grfica
plt.text(15, 0.12, f"$\mu$ = {mu}, $\sigma$ = {sigma}", fontsize=14, bbox=dict(facecolor='white', alpha=0.6))

# Permitir seleccionar un nivel de corriente y graficarlo
selected_current = 10  # Se puede cambiar a cualquier valor deseado
pdf_selected = lognorm.pdf(selected_current, sigma, scale=np.exp(mu))
cdf_selected = lognorm.cdf(selected_current, sigma, scale=np.exp(mu))
ax1.axvline(selected_current, color='blue', linestyle='--', label=f"Corriente = {selected_current} kA")
ax1.scatter([selected_current], [pdf_selected], color='blue', zorder=3)
ax2.scatter([selected_current], [cdf_selected], color='blue', zorder=3)
ax1.text(selected_current, pdf_selected, f"{pdf_selected:.4f}", fontsize=12, color='blue', verticalalignment='bottom')
ax2.text(selected_current, cdf_selected, f"{cdf_selected:.4f}", fontsize=12, color='blue', verticalalignment='bottom')
ax1.legend(fontsize=12)
ax2.legend(fontsize=12)

# Mostrar la grfica
plt.show()
```
